```python
from ultralytics import YOLO
import os
```

```python
# Load a model
model = YOLO('../models/bbox_cut/atpt_5_best.pt')
```

```python
# Dataset directory
data_dir = '../datasets/Train Numbers.v4-black-resize-no-contrast.yolov11'
```

```python
# Prepare test images
test_dir = f"{data_dir}/test/images"
images = [f'{test_dir}/{img}' for img in os.listdir(test_dir) if os.path.isfile(f'{test_dir}/{img}')]
print(images)

# Create test validation config
if not os.path.exists(f"{data_dir}/test.yaml"):
    lines = open(f"{data_dir}/data.yaml").readlines()
    for i, line in enumerate(lines):
        if line.startswith("val:"):
            lines[i] = "val: ../test/images\n"
            break
    open(f"{data_dir}/test.yaml", "w").writelines(lines)
    print("Created test validation config")
```

```python
metrics = model.val(data=os.path.abspath("../datasets/Train Numbers.v4-black-resize-no-contrast.yolov11/test.yaml"), batch=2)
```

```python
# Run batched inference on a list of images
results = model(images, batch=2)  # return a list of Results objects

# Process results list
for result in results:
    boxes = result.boxes  # Boxes object for bounding box outputs
    masks = result.masks  # Masks object for segmentation masks outputs
    probs = result.probs  # Probs object for classification outputs
    obb = result.obb  # Oriented boxes object for OBB outputs
    print(boxes, masks, probs, obb)
    result.save(filename=f"atpt_5_results/{result.path.rsplit('/', 1)[1]}")  # save to disk
```
